```python
import pandas as pd
df = pd.read_excel("New jobready_career_skill_dataset_1000_rows_v2.xlsx")
df.head()
```

```python
def advanced_gap(user_profile, career_skills):
    gaps = {}

    level_map = {"Beginner":1, "Intermediate":2, "Advanced":3}

    for career, skills in career_skills.items():
        missing = []
        weak = []

        for skill, weight, threshold, level in skills:

            if skill not in user_profile["skills"]:
                missing.append(skill)
            else:
                if level_map[user_profile["level"]] < level_map[level]:
                    weak.append(skill)

        gaps[career] = {
            "missing_skills": missing,
            "skills_to_improve": weak
        }

    return gaps
```
